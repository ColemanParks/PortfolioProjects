```python
#Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts plot configuration



#Read in the data

df = pd.read_csv(r'C:\Users\Colem\Downloads\movies.csv')
```

```python
#First Glimpse of Data 

df.head()
```

```python
#Reconfiguring View for Max rows

pd.set_option('display.max_rows', 1000)
```

```python
#Checking for omitted data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
#Rather than spend excess time filling in the missing data, I'm deleting any rows with NULL values

df = df.dropna(how='any',axis=0) 
```

```python
# Checking types of data in columns

df.dtypes
```

```python
#Reformatting data

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#Fixing Year of Release



df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df.head()
```

```python
#Dropping Duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.head()
```

```python
# Testing correlation between budget and gross revenue. 
# I predict a super positive correlation between the two due to:
# Star Power, Higher Marketing Budgets, and more access to better resources 
```

```python
# First, a Scatter Plot to test for any visual correlation between Budget and Gross Revenue

plt.scatter(x=df['budget'],y=df['gross'])
plt.show()
```

```python
# Plot Budget vs Gross

sns.regplot(x='budget',y='gross',data=df, scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
# Another view of all correlations with no visual aspect using the Pearson method

df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Previous Heatmap omits non-numeric qualities such as company or genre. 
# Now I'll numerize those categories and make a more detailed heatmap
```

```python
#Numerizing all values

df_numerized = df

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()

```
